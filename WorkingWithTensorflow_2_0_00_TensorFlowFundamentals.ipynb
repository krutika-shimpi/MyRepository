Create a vector, scalar, matrix and tensor with values of your choosing using tf.constant().

```python
# Scalar
scalar = tf.constant(6)

# Vector
vector = tf.constant([2,4,3,5,6,3])

# Matrix
matrix = tf.constant([[2,3,4],
                      [8,6,7],
                      [9,0,8]])

# Tensor
tensor = tf.constant([[[1,2,3,4],
                       [2,3,4,5]],
                      [[4,5,6,7],
                       [4,2,5,6]],
                      [[7,4,6,8],
                       [1,5,2,0]]])
```

Find the shape, rank and size of the tensors you created in 1.

```python
# Find the shape of the tensors
print(f'The shapes of the tensors are:\nScalar : {scalar.shape}\nVector: {vector.shape}\nMatrix : {matrix.shape}\nTensor : {tensor.shape}')

# Rank of the tensors
print(f'\nThe rank of the tensors are:\nScalar : {tf.rank(scalar)}\nVector: {tf.rank(vector)}\nMatrix : {tf.rank(matrix)}\nTensor : {tf.rank(tensor)}')

# Size of the tensors
print(f'\nThe sizes of the tensors are:\nScalar : {tf.size(scalar)}\nVector: {tf.size(vector)}\nMatrix : {tf.size(matrix)}\nTensor : {tf.size(tensor)}')
```

Create two tensors containing random values between 0 and 1 with shape [5, 300].

```python
# Creating random tensors
tensor_1 = tf.random.normal(shape = (5, 300))
tensor_2 = tf.random.normal(shape = (5, 300))
```

Multiply the two tensors you created in 3 using matrix multiplication.


```python
# Element wise multiplication tensor_1 * tensor_2
tensor_mul = tf.matmul(tensor_1, tf.transpose(tensor_2))
tensor_mul
```

```python
tensor_1 * tensor_2
```

Multiply the two tensors you created in 3 using dot product.

```python
# The dot product
tf.tensordot(tensor_1, tf.transpose(tensor_2), axes = 1)
```

```python
# A @ B
tensor_1 @ tf.transpose(tensor_2)
```

Create a tensor with random values between 0 and 1 with shape [224, 224, 3].

```python
# Create a tensor with random values between 0 and 1 with shape [224, 224, 3].
tensor_3 = tf.random.normal(shape = (224, 224, 3))
tensor_3.shape
```

Find the min and max values of the tensor you created in 6 along the first axis.

```python
# Find the min and max values of the tensor along the first axis.
min_values = tf.reduce_min(tensor_3, axis = 0)
max_values = tf.reduce_max(tensor_3, axis = 0)
print(f"Min values:\n{min_values}")
print(f"\n\nMax values:\n{max_values}")
```
